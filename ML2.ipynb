```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D']
values = [40, 30, 20, 10]

def simple_pie_chart():
    plt.figure()
    plt.pie(values, labels=categories, autopct='%1.1f%%')  # Fixed typo in autopct
    plt.title("Pie Chart")
    plt.show()

simple_pie_chart()
```

```python
simple_pie_chart()
```

```python
def simple_pareto_chart():
    sorted_pairs = sorted(zip(values, categories), reverse=True)
    sorted_values, sorted_categories = zip(*sorted_pairs)
    cummulative=  np.cumsum(sorted_values) / sum(sorted_values) * 100
    fig, axl = plt.subplots()

    axl.bar(sorted_categories, sorted_values)
    ax2 = axl.twinx()
    ax2.plot(sorted_categories, cummulative, color = 'red', marker = 'o')
    plt.title('pareto chart')
    axl.set_xlabel('categories')
    axl.set_ylabel('values')
    ax2.set_ylabel('cummulative %')
    
    plt.show()
```

```python
simple_pareto_chart()
```
