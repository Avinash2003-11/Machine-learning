```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D']
values = [40, 30, 20, 10]

def simple_pie_chart():
    plt.figure()
    plt.pie(values, labels=categories, autopct='%1.1f%%')  # Fixed typo in autopct
    plt.title("Pie Chart")
    plt.show()

simple_pie_chart()
```

```python
simple_pie_chart()
```

```python
def simple_pareto_chart():
    sorted_pairs = sorted(zip(values, categories), reverse=True)
    sorted_values, sorted_categories = zip(*sorted_pairs)
    cummulative=  np.cumsum(sorted_values) / sum(sorted_values) * 100
    fig, axl = plt.subplots()

    axl.bar(sorted_categories, sorted_values)
    ax2 = axl.twinx()
    ax2.plot(sorted_categories, cummulative, color = 'red', marker = 'o')
    plt.title('pareto chart')
    axl.set_xlabel('categories')
    axl.set_ylabel('values')
    ax2.set_ylabel('cummulative %')
    
    plt.show()
```

```python
simple_pareto_chart()
```

```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

df = pd.read_csv('./data/TSLA.csv')
df.head
```

```python
df.columns
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
x = df[['High', 'Low', 'Open', 'Volume']]
y = df['Close']
x.shape
```

```python
y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
print(model.coef_)
```

```python
print(model.intercept_)
```

```python
predicted = model.predict(x_test)
print(predicted)
```

```python
df1 = pd.DataFrame({'Actual': y_test.flatten(), 'predicted' : predicted.flatten()})
```

```python
print(y_test.shape, predicted.shape)
```

```python
df1 = pd.DataFrame({'Actual': y_test.values, 'Predicted': predicted.flatten()})
```

```python
df1
```

```python
import matplotlib.pyplot as plt
import numpy as np
df_sample = df1.head(10)
x = np.arange(len(df_sample))  
width = 0.4

plt.figure(figsize=(8, 5))
plt.bar(x - width/2, df_sample["Actual"], width=width, label="Actual", color="blue")
plt.bar(x + width/2, df_sample["Predicted"], width=width, label="Predicted", color="orange")
plt.xlabel('index')
plt.ylabel('values')
plt.title('Actual vs predicted - bar plot')
plt.xticks(ticks=x, labels=df_sample.index)
plt.legend()
plt.show()
```
