```python
df['Age']  = df['Age'].fillna(df['Age'].mean())
```

```python
df['Monthly_Usage'] = df['Monthly_Usage'].fillna(df['Monthly_Usage'].mean())
df['Churn'] = df['Churn'].fillna(df['Churn'].mean())
df['Customer_Support_Calls'] = df['Customer_Support_Calls'].fillna(df['Customer_Support_Calls'].mean())
```

```python
df['Subscription_Type'] = df['Subscription_Type'].fillna(df['Subscription_Type'].mode()[0])
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoded_type = encoder.fit_transform(df[['Subscription_Type']]).toarray()
```

```python
encoded_df = pd.DataFrame(encoded_type, columns=encoder.get_feature_names_out(['Subscription_Type']))
```

```python
concat = pd.concat([df, encoded_df], axis=1)
concat.head
```

```python
concat = concat.drop('Subscription_Type', axis=1)
concat.head
```

```python
x = concat.drop('Churn', axis=1)
y = concat['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42 )
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
scaler.fit(x_train, y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score, precision_score, recall_score
for k in range(1, 21, 2):  
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"k={k}, Accuracy={acc:.2f}")
```

```python
k = 13
model = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
model.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test)
print(prediction)
```

```python
accuracy = accuracy_score(y_test, prediction)
precision = precision_score(y_test, prediction)  
recall = recall_score(y_test, prediction)  
f1 = f1_score(y_test, prediction)  
confu_mat = confusion_matrix(y_test, prediction)
report = classification_report(y_test, prediction)
```

```python
print('model accuracy:', accuracy)
print('precision_score:', precision)
print('recall_score:', recall)
print('f1_score:', f1)
print('confusion_matrix:', confu_mat)
print('model report', report)
```
