```python
x = df.drop('fake', axis=1)
y = df['fake']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
scaler.fit(x_train, y_train)
```

```python
for k in range(1, 21, 2):  
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"k={k}, Accuracy={acc:.2f}")
```

```python
k = 13
model = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
model.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test)
print(prediction)
```

```python
from sklearn.metrics import accuracy_score ,confusion_matrix, recall_score,  precision_score, classification_report, f1_score
accuracy = accuracy_score(y_test, prediction)
precision = precision_score(y_test, prediction)  
recall = recall_score(y_test, prediction)  
f1 = f1_score(y_test, prediction)  
confu_mat = confusion_matrix(y_test, prediction)
report = classification_report(y_test, prediction)
```

```python
print('model accuracy:', accuracy)
print('precision_score:', precision)
print('recall_score:', recall)
print('f1_score:', f1)
print('confusion_matrix:', confu_mat)
print('model report', report)
```
