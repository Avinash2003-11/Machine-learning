```python
import pandas as pd
df = pd.read_csv('./data/decision_tree_dataset.csv')
df.head(10)
```

```python
df.isnull().sum()
```

```python
df['Loan_Approval'].value_counts()
```

```python
x = df.drop('Loan_Approval', axis=1)
y = df['Loan_Approval']
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = DecisionTreeClassifier(criterion='gini', max_depth=12, min_samples_split=6, random_state=42)
model.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test)
print(prediction)
```

```python
accuracy = accuracy_score(y_test, prediction)
print('model accuracy:', accuracy*100)
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test,y_test
            )
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
plot_tree(model, feature_names=['Age', 'Income', 'Credit_Score', 'Car_Ownership'], 
          class_names=['Not Approved', 'Approved'], filled=True)
plt.show()
```
