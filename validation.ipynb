```python
def analyse(numbers):
    avg = sum(numbers) / len(numbers)
    high = max(numbers)
    low = min(numbers)
    tot = sum(numbers)

    return avg, high, low, tot

a = [10, 20, 15, 25, 30]
analyse(a)
```

```python
import numpy as np

class TemperatureTracker:
    def __init__(self):
        self.temperatures = []

    def record_temperature(self, temp):
        
        self.temperatures.append(temp)

    def get_hottest_day(self):
        
        if not self.temperatures:
            return None
        return max(self.temperatures)

    def get_coldest_day(self):
        
        if not self.temperatures:
            return None
        return min(self.temperatures)

    def get_average_temperature(self):
       
        if not self.temperatures:
            return None
        return np.mean(self.temperatures)


tracker = TemperatureTracker()


daily_temps = [25, 30, 28, 35, 27, 31, 29]
for temp in daily_temps:
    tracker.record_temperature(temp)

print(f"Hottest Temperature: {tracker.get_hottest_day()}C")
print(f"Coldest Temperature: {tracker.get_coldest_day()}C")
print(f"Average Temperature: {tracker.get_average_temperature():.2f}C")
```

```python
import numpy as np


expenses = {
    'January': {'rent': 1000, 'food': 300, 'utilities': 150},
    'February': {'rent': 1000, 'food': 280, 'utilities': 170},
    'March': {'rent': 1000, 'food': 320, 'utilities': 160}
}


total_expenses_per_month = {month: sum(categories.values()) for month, categories in expenses.items()}


category_totals = {}
for month_data in expenses.values():
    for category, amount in month_data.items():
        category_totals[category] = category_totals.get(category, []) + [amount]

average_category_costs = {category: np.mean(costs) for category, costs in category_totals.items()}
highest_avg_category = max(average_category_costs, key=average_category_costs.get)


highest_spending_month = max(total_expenses_per_month, key=total_expenses_per_month.get)


print("Total Expenses per Month:", total_expenses_per_month)
print(f"Category with Highest Average Cost: {highest_avg_category} (${average_category_costs[highest_avg_category]:.2f})")
print(f"Month with Highest Total Spending: {highest_spending_month}")
```

```python
import numpy as np


inventory = [
    {'product': 'A', 'price': 10, 'quantity': 5},
    {'product': 'B', 'price': 20, 'quantity': 3},
    {'product': 'C', 'price': 15, 'quantity': 4}
]


total_inventory_value = sum(item['price'] * item['quantity'] for item in inventory)


product_values = {item['product']: item['price'] * item['quantity'] for item in inventory}
highest_value_product = max(product_values, key=product_values.get)


average_price = np.mean([item['price'] for item in inventory])

# Print results
print(f"Total Inventory Value: ${total_inventory_value}")
print(f"Product with Highest Total Value: {highest_value_product} (${product_values[highest_value_product]})")
print(f"Average Price per Product: ${average_price:.2f}")
```
