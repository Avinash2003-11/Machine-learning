```python
#bar plot
import matplotlib.pyplot as plt
import numpy as np
product = ['laptop', 'phone', 'tablet', 'watch', 'headphones']
sales = [50, 40, 60, 40, 45]
cl  =['red', 'green', 'blue', 'yellow', 'grey']

plt.figure(figsize=(5,4))

bars = plt.bar(product, sales, color = cl)
plt.bar_label(bars, labels=sales, label_type='edge', fontsize=10, padding=1)
plt.xlabel("Products")
plt.ylabel("sales")
plt.title("product sale distribution")
plt.show()
```

```python
#line plot
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

city1 = [30, 32, 31, 29, 28, 27, 26]
city2 = [22, 24, 23, 25, 26, 27, 28]
city3 = [15, 16, 14, 13, 14, 15, 16]

error  =np.array([1.5] * len(days))

plt.plot(days, city1, linestyle='-', marker='o', color='r', label="City A")
plt.fill_between(days, np.array(city1) - error, np.array(city1) + error, color = 'r', alpha = 0.2)

plt.plot(days, city2, linestyle='--', marker='*', color='g', label="City B", markersize = 9)
plt.plot(days, city3, linestyle='-.', marker='^', color='b', label="City C")

plt.xlabel("Days of the Week")
plt.ylabel("Temperature (C)")
plt.title("Weekly Temperature Data for Three Cities")
plt.legend()

plt.grid(True, linestyle = '--', alpha = 0.6)
plt.show()
```

```python
#histogram
import seaborn as sns
np.random.seed(42)  
scores = np.random.normal(loc=100, scale=15, size=1000)


mean_score = np.mean(scores)
std_score = np.std(scores)


plt.figure(figsize=(8, 5))
sns.histplot(scores, bins=30, kde=True, color='blue', edgecolor='black', alpha=0.6)


plt.axvline(mean_score, color='red', linestyle='--', label=f"Mean: {mean_score:.2f}")
plt.axvline(mean_score - std_score, color='green', linestyle=':', label=f"Mean - 1: {mean_score - std_score:.2f}")
plt.axvline(mean_score + std_score, color='green', linestyle=':', label=f"Mean + 1: {mean_score + std_score:.2f}")


plt.xlabel("Test Scores")
plt.ylabel("Frequency")
plt.title("Distribution of Test Scores")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)

plt.show()
```

```python
#subplots
categories = ['A', 'B', 'C', 'D']
values = [20, 40, 50, 60 ]
plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
plt.bar(categories, values)
plt.xlabel("categories")
plt.ylabel("values")
plt.title("bar plot")

plt.subplot(2,2,2)
x_values1 = np.linspace(0, 2 * np.pi, 100)  
sine = np.sin(x_values1)
plt.plot(x_values1, sine, marker='o', linestyle='-', color='red', label='Sine Wave')
plt.title("Line Plot(sine wave)")

plt.subplot(2,2,3)
plt.scatter(categories, values)
plt.xlabel("categories")
plt.ylabel("values")
plt.title("scatter plot")

plt.subplot(2,2,4)
plt.hist(values)
plt.title("histogram plot")

plt.suptitle("subplots")
plt.tight_layout()
plt.show()



```

```python
#advances scatter plot
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(0)
x = np.linspace(0, 10, 100)
y = 2 * x + 1 + np.random.normal(0, 2, size=100)  
z = np.random.rand(100)  
size = np.random.randint(20, 200, size=100) 


slope, intercept = np.polyfit(x, y, 1)
y_trend = slope * x + intercept


plt.figure(figsize=(10, 6))
scatter = plt.scatter(x, y, c=z, s=size, cmap='viridis', alpha=0.6, edgecolors="w", label='Data Points')


plt.colorbar(scatter, label='Color based on Z')


plt.plot(x, y_trend, color='red', linestyle='--', label='Trend Line')


plt.xlabel('X Variable')
plt.ylabel('Y Variable')
plt.title('Advanced Scatter Plot with Multiple Variables')
plt.legend()

# Add grid lines
plt.grid(True)

# Show the plot
plt.show()

```
